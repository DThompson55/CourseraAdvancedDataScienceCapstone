# Weather Visualization

In this data exploration step we look at weather patterns across the 4 year period and look for patterns.

The csv files were created in earlier steps.

```python
import pandas as pd
import numpy as np

import pylab as pl
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt

from sklearn.metrics import jaccard_similarity_score

#df_joined = pd.read_csv("~/Projects/CourseraCapstone/joined.csv")
```

```python
df_incidents = pd.read_csv("csvs/top_ten_311.csv", dtype={5: 'str', 6:'str'}) 
#df_incidents.drop(axis=1, columns=['Unique Key'], inplace=True)
df_incidents.set_index('Date Only', inplace=True)

df_weather = pd.read_csv("csvs/Weather_One_Shot.csv", index_col=0)
df_weather['Date Only'] = df_weather['DATE'].apply(lambda x: x[0:10])
df_weather.set_index('Date Only', inplace=True)
df_weather.drop(axis=1, columns=['DATE','Sunrise','Sunset'], inplace=True)

#df_joined = df_weather.join(df_incidents)
```

```python
pd.DataFrame(df_incidents.groupby(['Date Only', 'Complaint Type']).count()).to_csv('grouped_top_ten.csv')
```

```python
df_ten = pd.read_csv("csvs/grouped_top_ten.csv", parse_dates=[0], usecols=[0,1,2])
df_ten.rename(index=str, columns={"Unique Key": "Count"}, inplace=True)
df_ten.head(10)
```

## Blocked Driveway

A simple plot of the number of blocked driveway complaints each day for the 4+ year period.  Some spikes! What happened there?

```python
x = df_ten['Complaint Type']=='Blocked Driveway'

s = df_ten['Count'][x]

#print (df_incidents[x])
t = np.arange(0, df_ten.shape[0], 1)

fig, ax = plt.subplots()
ax.plot(s)

ax.set(xlabel='days', ylabel='Number of Calls',
       title='Blocked Driveway')
ax.grid()

#fig.savefig("311_calls.png")
plt.show()
```
