```python
import pandas as pd
import numpy as np

import pylab as pl
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt

from sklearn.metrics import jaccard_similarity_score

#df_joined = pd.read_csv("~/Projects/CourseraCapstone/joined.csv")
```

```python
df_weather = pd.read_csv("~/Projects/CourseraCapstone/Weather_One_Shot.csv", usecols=[1,4,6,7,8,9])
df_weather['Date Only'] = df_weather['DATE'].apply(lambda x: x[0:10])
#df_weather.set_index('Date Only', inplace=True)
#df_weather.drop(axis=1, columns=['DATE','Sunrise','Sunset'], inplace=True)

df_ten = pd.read_csv("~/Projects/CourseraCapstone/grouped_top_ten.csv", usecols=[0,1,2])
df_ten.rename(index=str, columns={"Unique Key": "Count"}, inplace=True)
df_ten.head(10)
```

```python
df_weather.head()
```

```python
print (df_ten['Date Only'].dtype, df_weather['Date Only'].dtype)
```

```python
df_joined = pd.merge(df_ten, df_weather)
df_joined['Complaint ID'] = df_joined['Complaint Type']
df_joined['Complaint ID'] = df_joined['Complaint ID'].map({'Blocked Driveway': 1,'HEAT/HOT WATER': 2,'Illegal Parking': 3,'Noise - Residential': 4,'Noise - Street/Sidewalk': 5,'Street Condition': 6,'Street Light Condition': 7,'UNSANITARY CONDITION': 8,'Water System': 9,'Request Large Bulky Item Collection': 10})

df_joined['dow'] = pd.to_datetime(df_joined['DATE']).dt.dayofweek


df_joined.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plotWEWD(complaint):
    a = (df_joined['Complaint Type']==complaint) & (~((df_joined['dow']==6) | (df_joined['dow']==5)))
    b = (df_joined['Complaint Type']==complaint) & (((df_joined['dow']==6) | (df_joined['dow']==5)))

    m1 = pd.DataFrame()
    m1['Temperature'] = df_joined['DailyAverageDryBulbTemperature'][a]
    m1['Weekday'] = df_joined['Count'][a]

    m2 = pd.DataFrame()
    m2['Temperature'] = df_joined['DailyAverageDryBulbTemperature'][b]
    m2['Weekend'] = df_joined['Count'][b]


    fig, ax = plt.subplots()
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    ax.plot(m1['Temperature'], m1['Weekday'], marker='o', linestyle='', ms=2)
    ax.plot(m2['Temperature'], m2['Weekend'], marker='o', linestyle='', ms=2)

    ax.legend()

    plt.show()
    
def plotWDWE(complaint):
    a = (df_joined['Complaint Type']==complaint) & (~((df_joined['dow']==6) | (df_joined['dow']==5)))
    b = (df_joined['Complaint Type']==complaint) & (((df_joined['dow']==6) | (df_joined['dow']==5)))

    m1 = pd.DataFrame()
    m1['Temperature'] = df_joined['DailyAverageDryBulbTemperature'][a]
    m1['Weekday'] = df_joined['Count'][a]

    m2 = pd.DataFrame()
    m2['Temperature'] = df_joined['DailyAverageDryBulbTemperature'][b]
    m2['Weekend'] = df_joined['Count'][b]


    fig, ax = plt.subplots()
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    ax.plot(m2['Temperature'], m2['Weekend'], marker='o', linestyle='', ms=2)
    ax.plot(m1['Temperature'], m1['Weekday'], marker='o', linestyle='', ms=2)

    ax.legend()

    plt.show()    
    
def plotTwo(complaint1, complaint2):    
    a = df_joined['Complaint Type']==complaint1
    b = df_joined['Complaint Type']==complaint2

    m1 = pd.DataFrame()
    m2 = pd.DataFrame()

    m1['A'] = df_joined['Count'][a]
    m2['B'] = df_joined['Count'][b]

    m1['T1']     = df_joined['DailyAverageDryBulbTemperature'][a]
    m2['T2']     = df_joined['DailyAverageDryBulbTemperature'][b]

    #print (m2['T2'])


    fig, ax = plt.subplots()
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    ax.plot(m1['T1'], m1['A'], marker='o', linestyle='', ms=2)
    ax.plot(m2['T2'], m2['B'], marker='o', linestyle='', ms=2)

    ax.legend()

    plt.show()

def plotOne(complaint):
    a = df_joined['Complaint Type']==complaint

    m = pd.DataFrame()
    m['Temperature'] = df_joined['DailyAverageDryBulbTemperature'][a]
    m['Number of Calls'] = df_joined['Count'][a]

    m.plot.scatter(x='Temperature',y='Number of Calls')


def plotOneAvg(complaint):
    a = df_joined['Complaint Type']==complaint

    m = pd.DataFrame()
    m['Temperature'] = df_joined['DailyAverageDryBulbTemperature'][a]
    m['Number of Calls'] = df_joined['Count'][a]
    
    s = pd.DataFrame(m.groupby('Temperature').mean())
    s.reset_index(inplace=True)
    
#    print(m.groupby('Temperature').mean())
    s.plot.scatter(x='Temperature',y='Number of Calls')

plotOneAvg('HEAT/HOT WATER')
    
```
